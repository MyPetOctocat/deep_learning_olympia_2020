```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import math
import os
```

```python
#version = "normal"
version = "filter"



weights = "yolov4-{}.weights".format(version)
cfg = "yolov4-{}.cfg".format(version)

#directory = r'/home/dl4/Task2c/test/filter_imgs/'
directory = r'/home/dl4/Task2c/test/normal_imgs/Test/'
#directory = r'/home/dl4/Task2/Video_Frames/_tigfCJFLZg_00435/'

result = [0, 0]

net = cv2.dnn.readNet(weights, cfg)    # yolo Modell laden
```

```python
for filename in os.listdir(directory):
    path = directory + filename
    print(path)
    
    img = cv2.imread(path)
    height, width, _ = img.shape
    
    
    blob = cv2.dnn.blobFromImage(img, 1/255, (416, 416), (0,0,0), swapRB=True, crop=False)
    
    # Frame wird durch das NN propagiert
    net.setInput(blob) 
    output_layers_names = net.getUnconnectedOutLayersNames()
    layerOutputs = net.forward(output_layers_names)
    

    
    # Variablen für die Detection werden initialisiert
    boxes = []
    confidences = []
    class_ids = []

    for output in layerOutputs:
        for detection in output:
            scores = detection[5:]
            class_id = np.argmax(scores)
            confidence = scores[class_id]
            if confidence > 0.5:
                center_x = int(detection[0]*width)
                center_y = int(detection[1]*height)
                w = int(detection[2]*width)
                h = int(detection[3]*height)

                x = int(center_x - w/2)
                y = int(center_y - h/2)

                boxes.append([x, y, w, h])
                confidences.append(float(confidence))
                class_ids.append(class_id)

    indices = cv2.dnn.NMSBoxes(boxes, confidences, 0.5, 0.4) #Non Max Supression, um endgültige Anzahl an erkannten Objekten zu bekommen
        
    if len(indices) > 0: # Wenn ein Objekt gefunden wurde
        result[1] += 1
    else:
        result[0] += 1

    
print("Watersplash erkannt:\t{}".format(result[0]))

print("Watersplash nicht erkannt:\t{}".format(result[1]))
```
