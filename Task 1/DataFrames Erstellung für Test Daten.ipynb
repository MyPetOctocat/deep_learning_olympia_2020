```python
#Laden der Videos und Speicherung der Bilder 
import cv2
import os
path="/home/dl4/Video_Frames_Test/video_data"
#schleife durchläuft alle videos
for i in range (105,130):
    #die Methode cv2.VideoCapture greift auf die einzelnen Videos zu
    vidcap = cv2.VideoCapture('/home/dl4/Videos_Test/_8Vy3dlHg2w_00'+str(i).zfill(3)+'.mp4')
    #Vicap.read() ob das Bild des Videos geladen ist und gibt die aktuelle Bilddatei zurück
    success, image = vidcap.read()
    count = 1
    #erstellt pro Video ein Ordner
    os.makedirs('Video_Frames_Test/video_data'+str(i))
    #Alle Bilder werden geladen und gespeichert

    while success:
        cv2.imwrite('Video_Frames_Test/video_data'+str(i)+'/image_'+str(count)+'_'+str(i)+'.jpg',image)
        success, image = vidcap.read()
        print('Saved image ', count)
        count += 1
```

```python
#Speicherung der gelabelten Test-Daten CSV-Datein in einem Dataframe
import pandas as pd
df = pd.read_csv("/home/dl4/Labels/Task1_Test_Labels_Videos.csv", sep=";")
print(df)
df = df.rename(columns={"ID" : "ID","NameOfFile" : "video", "Camera Position (Side/Front/Upwards/Downwards)" : "cam_pos"})
```

```python
print(df.shape)
for i in range(df.shape[0]):
    df["video"][i] = "video_data" + str(df["ID"][i])
print(df)
```

```python
#Hinzufügen der Kameraposition zu den Videonamen
directory = "/home/dl4/Video_Frames_Test/"

# Append cam position to folder name

#!! wenn vollständige excel da ist müssen die bearbeiteten files runtergenommen werden // oder if-abfrage

for filename in os.listdir(directory):
    if filename in df.video.values:
        cam = df.loc[df['video'] == filename]["cam_pos"]
        cam = cam.iloc[0]
        print(filename + "_" + cam)
        os.rename(directory + filename, directory + filename + "_" + cam)
```
