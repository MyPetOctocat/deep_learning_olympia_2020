# Graphen für die Präsentation

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

#Farbschema
color = [0.0059/30, 4.7370e-5, 7.0546e-6, 3.0094e-5]
gray = [0.0199/30, 0.0080/30, 1.7407e-4, 1.3781e-4]
res = ["28x28", "32x32", "64x64", "128x128"]
```

```python
x = res
y1 = color
y2 = gray

plt.plot(x, y1, "-r", label="color", marker="o")
plt.plot(x, y2, "-k", label="gray", marker="o")
plt.ylabel("validation loss")
plt.xlabel("resolution")
plt.legend(loc="upper right")
#plt.ylim(-1.5, 2.0)
plt.show()
```

```python
#Illustration Overfit

val_loss = [7.5607e-5, 7.0755e-6, 4.5745e-6, 2.6854e-6, 5.3884e-7, 1.2151e-7]
test_loss = [8.6630e-4/5, 6.2722e-4/5, 1.9309e-5, 2.7087e-5, 0.0054/30, 0.0083/30]
epochs = ["5", "10", "15", "20", "25", "30"]
```

```python
x = epochs
y1 = val_loss
y2 = test_loss

plt.plot(x, y1, "-b", label="validation loss", marker="o")
plt.plot(x, y2, "-y", label="test loss", marker="o")
plt.ylabel("loss")
plt.xlabel("epochs")
plt.legend(loc="upper center")
plt.show()
```
